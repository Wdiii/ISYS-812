### Import Video dataset

```python
US_vid = pd.read_csv('USvideos.csv')
GB_vid = pd.read_csv('GBvideos.csv')
```

### Import category dataset

```python
US_cat = pd.read_csv('US_category_id.csv')
GB_cat = pd.read_csv('US_category_id.csv')
```

### Merge US video dataset with category dataset for category column

```python
US_vid.category_id = US_vid.category_id.astype(str)
US_cat.items__id = US_cat.items__id.astype(str)

US_data = US_vid.merge(US_cat, left_on = 'category_id', right_on = 'items__id')

US_data.drop(['kind', 'etag', 'items__kind', 'items__etag', 'items__snippet__channelId', 'items__id', 'items__snippet__assignable'], axis = 1, inplace = True)

US_data.rename(columns = {'items__snippet__title':'category'}, inplace = True)

US_data
```

### Merge GB video dataset with category dataset for category column

```python
GB_vid.category_id = GB_vid.category_id.astype(str)
GB_cat.items__id = GB_cat.items__id.astype(str)

GB_data = GB_vid.merge(GB_cat, left_on = 'category_id', right_on = 'items__id')

GB_data.drop(['kind', 'etag', 'items__kind', 'items__etag', 'items__snippet__channelId', 'items__id', 'items__snippet__assignable'], axis = 1, inplace = True)

GB_data.rename(columns = {'items__snippet__title':'category'}, inplace = True)

GB_data
```

### Data cleansing, remove unnecessary columns, remove NaN rows.

#### US Data

```python
US_data.drop(['video_id', 'tags', 'thumbnail_link','description'], axis = 1, inplace = True)
US_data.drop(US_data[US_data['comments_disabled'] == True].index, inplace = True)
US_data.drop(US_data[US_data['ratings_disabled'] == True].index, inplace = True)
US_data.drop(US_data[US_data['video_error_or_removed'] == True].index, inplace = True)

US_data
```

```python
US_data['publish_time'] = US_data['publish_time'].apply(lambda x : x[0:10].replace('-','.'))

US_data['publish_time'] = pd.to_datetime(US_data['publish_time'], format = '%Y.%m.%d')

US_data['trending_date'] = pd.to_datetime(US_data['trending_date'], format = '%y.%d.%m')

US_data
```

#### GB Data

```python
GB_data.drop(['video_id', 'tags', 'thumbnail_link','description'], axis = 1, inplace = True)
GB_data.drop(GB_data[GB_data['comments_disabled'] == True].index, inplace = True)
GB_data.drop(GB_data[GB_data['ratings_disabled'] == True].index, inplace = True)
GB_data.drop(GB_data[GB_data['video_error_or_removed'] == True].index, inplace = True)

GB_data
```

#### Convert time data to datetime.

```python
GB_data['publish_time'] = GB_data['publish_time'].apply(lambda x : x[0:10].replace('-','.'))

GB_data['publish_time'] = pd.to_datetime(GB_data['publish_time'], format = '%Y.%m.%d')

GB_data['trending_date'] = pd.to_datetime(GB_data['trending_date'], format = '%y.%d.%m')

GB_data
```

```python
def count_words(input_list):
    count_dict = {}
    index = 0
    for title in input_list:
        word_list = re.split('\W+', title)
        for word in word_list:
            if word != "" and word.lower() not in stop_word_list:
                word = word.lower()
                count = count_dict.setdefault(word,0)
                count_dict[word] = count + 1
        index +=1
    return  count_dict
```

```python
def quick_sort(collection, low, high):
    if low >= high:
        return collection
    else:
        pivot = collection[low]  
        left = low
        right = high
        while left < right:
            while left < right and collection[right][1] <= pivot[1]:
                right -= 1   
            collection[left] = (collection[right][0],collection[right][1])
            while left < right and collection[left][1] >= pivot[1]:
                left +=1   
            collection[right] = (collection[left][0],collection[left][1])
        collection[right] = pivot    
        quick_sort(collection, low, left-1)    
        quick_sort(collection, left+1, high)    
        return collection
```

```python
import math
```

```python
def count_calculator(coeff, number) :
    temp_val = math.exp(coeff * number)
    final = (temp_val - 1) * 100
    returner = str(final) + '%'
    return returner

def categorical_calculator(coeff) :
    temp_val = math.exp(coeff)
    return temp_val
```

```python
print ('Results for USA: ')
print ('The Intercept value of the Poisson Regression model is ' + str(US_model_results['Intercept']))

for eachone in US_model_results['Categorical'] :
    print ("A change in ‘category’ from ‘Autos & Vehicles’ to " +
           "'" + str(eachone) + "' " + "causes a change in the View Count by a factor of " +
           str(categorical_calculator(US_model_results['Categorical'][eachone])))

for eachone in US_model_results['Count'] :
    print ("For an increase in the " + "'" + str(eachone) + "'" + " by 1, the View Count increases by a factor of: " +
           (count_calculator(US_model_results['Count'][eachone], 1)))
```
